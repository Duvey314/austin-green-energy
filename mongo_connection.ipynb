# Mongodb Connection
The Mongo database is fairly straigh forward to connect to. Before connecting to the database you need to install the pymongo driver in your environment. You might also need to install pymongo-srv (you can see this in the first part of the connection string "mongo+srv"). [This posting](https://stackoverflow.com/questions/52930341/pymongo-mongodbsrv-dnspython-must-be-installed-error) on it can help you install the package.

Once theses have been installed you can navigate to the mongodb atlas 'austin-green-energy' cluster on your web browser and select connect. You'll then select 'connect your application.' On the next screen select python and your version of python (this example uses python 3.6 or later) and it will generate the connection string. Copy the connection string and paste it below. Be sure to change the default database and include a config file with your username and password.


```python
# import dependencies
import config
import pymongo
import pandas as pd
import json
```

```python
# set string variables
DEFAULT_DATABASE = 'sample_airbnb'
USERNAME = config.USERNAME
PASSWORD = config.PASSWORD

#create connection to database
client = pymongo.MongoClient(f"mongodb+srv://{USERNAME}:{PASSWORD}@austin-green-energy.pwzpm.mongodb.net/{DEFAULT_DATABASE}?retryWrites=true&w=majority")
```

```python
# select datase
db = client.get_database('sample_airbnb')
#select collection in database
collection = db.listingsAndReviews

# pull collection into dataframe
df = pd.DataFrame(list(collection.find()))
df.head()
```

```python
# list all of the databases
for db in client.list_databases():
    print(db)
```

```python
# select database
db = client.get_database('wind_solar_data')
# select collection
collection = db.wind_data

data = pd.read_csv('.\Resources\Hackberry_Generation.csv')
data_json = json.loads(data.to_json(orient='records'))
collection.remove()
collection.insert(data_json)
```

```python
# pull collection into dataframe
df = pd.DataFrame(list(collection.find()))
df.head()
```
